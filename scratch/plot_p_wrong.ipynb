```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def calc_pwrong(m,s):
    return 0.5+0.5*scipy.special.erf(m/np.sqrt(2+2*s**2))
```

```python
n_samples = 1000
m=np.linspace(-10, 10.0, num=n_samples)
s=[0.2,1,2,3,5]
pwrong_all = np.zeros((np.size(s),n_samples))
for i in range(1000):
    for j in range(np.size(s)):
        pwrong_all[j,i] = calc_pwrong(m[i],s[j])
        rc('font',**{'family':'serif','serif':['Helvetica']})
rc('text', usetex=True)
plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots()
ax.set_xlabel(r'$m$')
ax.set_ylabel(r'$P_{wrong}(A>B)$')
ax.plot([-10,10],[0.5,0.5],color = 'black',  linestyle='dashed')
for j in range(np.size(s)):
    ax.plot(m,pwrong_all[j,:],label=r'$s=\enspace $'+str(s[j]))
ax.legend()
plt.grid()
fig.tight_layout()
fig.savefig('pwrong_vs_m.pdf')
plt.show()
```
